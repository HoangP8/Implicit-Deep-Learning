```python
from examples.im_rnn.netflix_stock import netflix_dataset
from examples.im_rnn.spiky_data import spiky_synthetic_dataset
from matplotlib import pyplot as plt
import numpy as np
```

```python
def visualize_train_test_split(x_train, x_test, y_train, y_test, dataset_name='Dataset', y_label=None):
    """
    Visualize the training and testing target values over time.

    Args:
        x_train (torch.Tensor): Training input sequences
        x_test (torch.Tensor): Testing input sequences
        y_train (torch.Tensor): Training target values
        y_test (torch.Tensor): Testing target values
        dataset_name (str): Name of the dataset
    """
    # Convert tensors to numpy arrays
    y_train_np = y_train.numpy().flatten()
    y_test_np = y_test.numpy().flatten()

    # Create time axes
    train_time = np.arange(len(y_train_np))
    test_time = np.arange(len(y_train_np), len(y_train_np) + len(y_test_np))

    plt.style.use('ggplot') 
    plt.rcParams['axes.grid'] = True  
    plt.rcParams['axes.edgecolor'] = 'black'
    plt.rcParams['grid.color'] = 'white' 
    plt.rcParams['grid.linestyle'] = '--'  
    plt.rcParams['grid.linewidth'] = 0.5
    plt.rcParams['axes.labelcolor'] = 'black'
    plt.rcParams['xtick.color'] = 'black' 
    plt.rcParams['ytick.color'] = 'black'  
    plt.rcParams['axes.labelpad'] = 8

    plt.figure(figsize=(6, 4))
    plt.plot(train_time, y_train_np, label='Train', color='blue')
    plt.plot(test_time, y_test_np, label='Test', color='red')
    plt.axvline(x=len(y_train_np), color='black', linestyle='--', label='Train/Test Split')
    plt.title(f"{dataset_name.capitalize()} Dataset", fontsize=12)
    plt.xlabel("Time Steps", fontsize=10)
    plt.ylabel(y_label, fontsize=10)
    plt.legend()
    plt.show()
```

```python
x_train, x_test, y_train, y_test = netflix_dataset(look_back=60)
print('x_train.shape = ',x_train.shape)
print('y_train.shape = ',y_train.shape)
print('x_test.shape = ',x_test.shape)
print('y_test.shape = ',y_test.shape)
visualize_train_test_split(x_train, x_test, y_train, y_test, dataset_name='netflix', y_label='Volatility')
```

```python
x_train, x_test, y_train, y_test = spiky_synthetic_dataset(look_back=60)
print('x_train.shape = ',x_train.shape)
print('y_train.shape = ',y_train.shape)
print('x_test.shape = ',x_test.shape)
print('y_test.shape = ',y_test.shape)    
visualize_train_test_split(x_train, x_test, y_train, y_test, dataset_name='spiky synthetic')
```
